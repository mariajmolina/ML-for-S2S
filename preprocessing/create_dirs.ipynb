### Creating directories to input new variables into

```python
import os
import numpy as np
import pandas as pd
from itertools import product
from config import parent_directory, start_date, end_date, variable_folder, starts
```

Hindcasts created from 1999-2019, January-December.

Files to be saved are of daily temporal resolution.

```python
d1 = pd.date_range(start_date, end_date, freq=starts)
```

Variable folder to be created.

```python
if os.path.exists(parent_directory):
    
    for yr, mo in product(np.unique(d1.strftime("%Y")), np.unique(d1.strftime("%m"))):
        
        new_directory = variable_folder+yr+"/"+mo
        path = os.path.join(parent_directory, new_directory)
    
        try: 
            os.makedirs(path, exist_ok = True) 
            print("Directory '%s' created successfully" % new_directory) 
        except OSError as error: 
            print("Directory '%s' cannot be created" % new_directory)
    
if not os.path.exists(parent_directory):
    print('Parent directory does not exist.')
```
